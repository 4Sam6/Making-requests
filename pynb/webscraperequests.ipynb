```python
import requests
import pandas as pd

# API request
url = "https://api.worldbank.org/v2/country/USA;DEU;CHN;AUS;NGA;BRA/indicator/NY.GNP.PCAP.PP.CD?format=json&per_page=100"
response = requests.get(url)

# Check if the response is valid
if response.status_code == 200:
    json_request = response.json()

    # Print the structure of the response to inspect the data
    if isinstance(json_request, list) and len(json_request) > 1:
        print(json_request[1])  # Print data section

        # Convert to DataFrame
        df = pd.DataFrame(json_request[1])
        pd.set_option("display.max_columns", None)  # Show all columns
        print(df.head())  # Inspect the data
    else:
        print("No data found in the response.")
        df = pd.DataFrame()
else:
    print(f"Error: {response.status_code}")
```

```python
country_codes_requested = ["USA", "DEU", "CHN", "AUS", "NGA", "BRA"]
all_data = []
for country in country_codes_requested:
    country_url = f"https://api.worldbank.org/v2/country/{country}/indicator/NY.GNP.PCAP.PP.CD?format=json"
    response = requests.get(country_url).json()
    if len(response) > 1 and response[1] is not None:
        all_data.extend(response[1])

df = pd.DataFrame(all_data)
```
