```python
import requests
import pandas as pd

# API request
url = "https://api.worldbank.org/v2/country/USA;DEU;CHN;AUS;NGA;BRA/indicator/NY.GNP.PCAP.PP.CD?format=json"
response = requests.get(url)
json_request = response.json()

# Extract the data
if len(json_request) > 1:
    df = pd.DataFrame(json_request[1])  # Extract relevant part
else:
    df = pd.DataFrame()

# Keep only relevant columns
df = df[['country', 'countryiso3code', 'date', 'value']]

# Convert 'country' column from a dictionary to a string
df['country'] = df['country'].apply(lambda x: x['value'] if isinstance(x, dict) else x)

# Drop rows with missing GNI values
df = df.dropna(subset=['value'])
df = df.drop(columns=['countryiso3code'])
# Convert 'date' to an integer
df['date'] = df['date'].astype(int)
df['value'] = df['value'].astype(int)
# Sort by country and year
df = df.sort_values(by=['country', 'date'])

```
