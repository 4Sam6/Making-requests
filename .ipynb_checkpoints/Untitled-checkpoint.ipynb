```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import wbdata
import requests
import pandas as pd
from datetime import datetime
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# This will give you a variable that refers to the  indicator of total population of each country, 
# it is a wbdata.client.SearchResult, with this we can 
indicator = wbdata.get_indicators('SP.POP.TOTL')

# Fetch data from the World Bank API
df = wbdata.get_dataframe({"SP.POP.TOTL" : "Population, total"}, country=["NGA","USA","DEU", "BRA" ], date = ( "1960", "2025") )

# This is cleaning the data so it is ready to be used 

df = df.reset_index()
df = df.rename(columns={"date": "Years", "country": "Country", "Population, total": "Population"})
df["Years"] = df["Years"].astype(int)
df["Population"] = pd.to_numeric(df["Population"], errors="coerce")
df["Population"] = df["Population"].astype(int)
df = df.dropna()
df = df.sort_values(by=["Country", "Years"])
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib.offsetbox import OffsetImage, AnnotationBbox  

# Sample data
countries = ["United States", "Nigeria", "Germany", "Brazil"]
population = [331, 206, 83, 213]  # Population in millions

# Image paths (Ensure these files exist in the correct directory)
image_paths = ["United_States.png", "Nigeria.png", "Germany.png", "Brazil.png"]

# Create figure and axis
fig, ax = plt.subplots(figsize=(8, 5))  

# Create horizontal bar chart
bars = ax.barh(df["Country"], df["Population"], color='skyblue', alpha=0.7)  # Adjust transparency

# Get maximum width (longest bar) for scaling images
max_width = max([bar.get_width() for bar in bars])

# Add images inside bars using OffsetImage & AnnotationBbox
for bar, img_path in zip(bars, image_paths):
    img = mpimg.imread(img_path)  # Load image

    # Dynamically adjust zoom based on bar width
    zoom_factor = bar.get_width() / max_width * 0.5  # Adjust as needed for flag size

    imagebox = OffsetImage(img, zoom=zoom_factor)  # Adjust zoom dynamically

    # Get the center of the bar
    x_center = bar.get_x() + bar.get_width() - (bar.get_width() * 0.1)  # Move flag slightly to the right
    y_center = bar.get_y() + bar.get_height() / 2  # Center vertically

    # Create annotation box (image inside the bar)
    ab = AnnotationBbox(imagebox, (x_center, y_center), frameon=False, xycoords="data", box_alignment=(1, 0.5))
    ax.add_artist(ab)

# Labels and title
ax.set_xlabel("Population (millions)")
ax.set_ylabel("Country")
ax.set_title("Population of Selected Countries with Flags")

# Adjust Y-axis ticks
ax.set_yticks(range(len(countries)))  
ax.set_yticklabels(countries)

# Show the plot
plt.show()
```
